## Tasks for the data

## 1. Which team won the most matches?

```python
df.head()
match_wins = df['match_winner'].value_counts()
sn.barplot(y = match_wins.index , x = match_wins.values , palette='viridis' )
plt.title("Most match win by team")
```

## 2. Toss decision trends

```python
sn.countplot(x = df['toss_decision'] ,hue = 'highscore', data = df , palette='rainbow')
plt.title('Toss decision trends')
```

## 3. Toss winner VS match winner

```python
count =df[df['toss_winner'] == df['match_winner']]['match_id'].count()
percentage  = (count * 100)/df.shape[0]
percentage.round(2)
```

## 4. How do team wins ? (Runs / Wickets)

```python
sn.countplot(x = df['won_by'] , palette='rainbow' )
plt.title('Winning side ')
```

## Key player performances

1. Most "player of the match" Awards

```python
count = df['player_of_the_match'].value_counts().head(10)
count
```

```python
sn.barplot(y = count.index , x = count.values , palette='mako')
plt.title("Top 10 Players with Man of the Match")
```

2. Top Scorers

```python
df.head()
```

```python
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending = False).head()
high
high.plot(kind = 'barh' )
plt.title('Top scorers')
```

10 best bowling Figures

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x : x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
```

```python
top_blowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending = False)
top_10 = top_blowlers.head(10)

sn.barplot(y = top_10.index , x = top_10.values , palette='rainbow' )
```
